```python
#find a number is prime
def is_prime(num):
    '''
    This function checks if a number is prime
    '''
    for n in range(2,num):
        if num % n == 0:
            print(num,'is not prime')
            break
    else:
        print(num,'is prime')
```

```python
is_prime(16)
```

```python
is_prime(7)
```

```python
#min_two_odds(3,5) --> 3
#min_two_odds(4,9) --> 9
def min_two_odds(n1,n2):
    if n1%2 !=0 and n2%2 !=0:
        return min(n1,n2)
    else:
        return max(n1,n2)
```

```python
min_two_odds(9,3)
```

```python
min_two_odds(9,2)
```

```python
#check if two-word string starts with the same letter
def two_words_letters(txt):
    wdlist = txt.split()
    word1 = wdlist[0]
    word2 = wdlist[1]
    return word1[0] == word2[0]
```

```python
#create_thirty(10,20) --> True
#create_thirty(5,30) --> True
#create_thirty(5,20) --> False
def create_thirty(num1,num2):
    if (num1 + num2) == 30:
        return True
    elif num1 == 30:
        return True
    elif num2 == 30:
        return True
    else:
        return False
```

```python
def caps_two_letters(word):
    if len(word) > 3:
        part1 = word[:3].capitalize()
        part2 = word[3:].capitalize()
        return part1 + part2
    else:
        return 'sorry, the word is too small'
```

```python
caps_two_letters('program')
```

```python
caps_two_letters('macdonald')
```

```python
caps_two_letters('cis')
```

```python
#Check if a list has number 5 next to number 5 somewhere in the list
#[3,4,5,5,10] -> true
# [3,4,5,6,5,8] -> false
def has_55(lst):
    for i in range(0,len(lst)-1):
        #if lst[i] == 5 and lst[i+1] == 5:
        if lst[i:i+2] == [5,5]:
            return True
        
    return False
```

```python
has_55([3,4,5,5,10])
```

```python
has_55([3,4,5,6,5,8])
```

```python
#check code 001 in a list
# [1,4,5,0,0,1,8] -> True
# [1,0,2,5,0,6,1,8] -> True
# [1,5,6,0,6,7,8,0] -> False

def check_code(lst):
    
    code = [0,0,1,'s']
    
    for num in lst:
        if num == code[0]:
            code.pop(0) #remove the number at index 0
            
    return len(code) == 1
```

```python
check_code([1,4,5,0,0,1,8])
```

```python
check_code([1,0,2,5,0,6,1,8])
```

```python
check_code([1,5,6,0,6,7,8,0])
```
